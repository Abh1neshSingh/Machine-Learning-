                                              Ada Boost Classifier

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier, AdaBoostRegressor
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,r2_score

import warnings
warnings.filterwarnings("ignore")
```

```python
X,y = make_classification(n_samples=1000,n_features=20,n_classes=2,random_state=1)
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=1)
```

```python
# now come model training part
```

```python
classifier = AdaBoostClassifier()
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
print("Accuracy Score ",accuracy_score(y_test,y_pred))
```

```python
print("classification_report\n",classification_report(y_test,y_pred))
```

```python
print("Confusion Matrix\n",confusion_matrix(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV # for Hyperparameter tuning 

param_grid = { "n_estimators": [50,100,200],
               "learning_rate":[0.001,0.1,1,1.5,2,2.5],
              "algorithm":["SAMME.R","SAMME"]
    
}

Ada = AdaBoostClassifier()
clf = GridSearchCV(estimator=Ada,param_grid=param_grid,cv=5,verbose=3)
```

```python
clf
```

```python
clf.fit(X_train,y_train)
```

```python
clf.best_score_
```

```python
clf.best_params_
```

```python
best_model=clf.best_estimator_
y_pred = best_model.predict(X_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import r2_score

import warnings
warnings.filterwarnings("ignore")
```

```python
X,y = make_regression(n_samples=1000,n_features=2,noise=10,random_state=1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=1)
```

```python
reg = AdaBoostRegressor()
```

```python
reg
```

```python
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
y_pred
```

```python
r2_score(y_test , y_pred)
```

```python
from sklearn.model_selection import GridSearchCV # for Hyperparameter tuning 

param_grid = { "n_estimators": [50,100,200],
               "learning_rate":[0.001,0.1,1,1.5,2,2.5],
              "loss":["linear","square","exponential"]
    
}

regg = AdaBoostRegressor()
reg = GridSearchCV(estimator=regg,param_grid=param_grid,cv=5,verbose=3)
```
