```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# first of all scaling is optional 
# what is scaling ?
# Scaling >> It transform the feature or varible 
# It bring the all the feature on same scale
```

```python
# why scaling >>
# 1)- To reduce to compuation time or in other word use for faster optimisation and
#     better interpretation
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
# Standardisation >>  is noting but Zscore
#  Zscore = x-mu/sigma
#  mu --   Mean
#  sigma -- std 

# use--->when you want to use your data , to between the  mu = 0 and std or sigma = 1
# then you use Standardisation 
```

```python

mean = np.mean(df['total_bill'])
```

```python
mean
```

```python
std = np.std(df['total_bill'])
```

```python
std
```

```python
std_data=[]
for i in list(df['total_bill']): # x is ----> i
    zscore =(i-mean)/std
    std_data.append(zscore)
```

```python
std_data
```

```python
# scaling is optional because it does'nt the distribution of data
plt.figure(figsize = (12, 5))
plt.subplot(1, 2, 1)
sns.distplot(df['total_bill'])
plt.title("Original Data of Total Bill")


plt.subplot(1, 2, 2)
sns.distplot(std_data)
plt.title("Standardised Data of Total Bill")
plt.show()
```

```python
# See Above the distribution of data is same 
# value is changed 
```
