```python
Class_0={'feature1':np.random.normal(0, 1,no_of_class_0),
 'feature2':np.random.normal(0, 1,no_of_class_0),
 'target':[0]*no_of_class_0
}
```

```python
Class_1=pd.DataFrame({'feature1':np.random.normal(3, 1,no_of_class_1),
 'feature2':np.random.normal(3, 1,no_of_class_1),
 'target':[1]*no_of_class_1
})
```

```python
# concatenat the both Class_0 And Class_1
```

```python
df=pd.concat([Class_0,Class_1]).reset_index(drop=True)
```

```python
# here come problem of class imbalance
```

```python
df_minority = df[df.target == 1]
df_majority = df[df.target == 0]
```

```python
df_majority
```

```python
# now >> oversampling  or upsampling >> increase the minority data to equivalent to majority data
```

```python
from sklearn.utils import resample
```

```python
df_minority_upsampled = resample(df_minority,replace=True,n_samples=len(df_majority),random_state=1)
```

```python
df_minority_upsampled
```

```python
df_minority_upsampled.shape
```

```python
df_minority_upsampled.duplicated().sum()
```

```python
df_upsampled = pd.concat([df_majority,df_minority_upsampled])
```

```python
df_upsampled
```

```python
df_upsampled.target.value_counts()
```

```python
from sklearn.datasets import make_classification 
# sklearn has  module called as datasets where you make classification 
# >>  make_classification is meathod or function >> you will use and start making a  dataset
```

```python
# so i used just  make_classification 
```

```python
x,y=make_classification( n_samples=1000, n_redundant=0 , n_features=2,n_clusters_per_class=1,weights=[0.90],random_state=1) 
# n_redundant means if any want any feature redundant >>no
# you want to two features >> n_feature
# for each class we have cluster of class0 and class1 >> n_cluster_per_class
# you want class have 90% data >> weights=[0.90]
# it give to two data so >> store in two variable >> x and y 
```

```python
#  this is  condition of class imbalance 
```

```python
# now >> what i am going to do
# I am going to actually  create a DataFrame for both x and y 
```

```python
df1= pd.DataFrame(x, columns = ['f1','f2'])
df2= pd.DataFrame(y, columns = ['target'])
df_final=pd.concat([df1,df2], axis=1)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
# make object of SMOTE()
over_sample= SMOTE()
```
