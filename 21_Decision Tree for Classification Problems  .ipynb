Decision Tree for Classication Problem

```python
import numpy as numpy 
import pandas  as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import load_iris
```

```python
data=load_iris()
df= pd.DataFrame(data.data,columns=data.feature_names)
df['target']=data.target
```

```python
df
```

```python
X=df.drop("target",axis=1)
y=df['target']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=1)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='entropy')
classifier
```

```python
classifier.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize =(12, 10))
tree.plot_tree(classifier, filled=True)
```

```python
classifier.predict(X_test)
```

```python
# post pruning 
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='entropy',max_depth=2)
classifier
classifier.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize =(12, 10))
tree.plot_tree(classifier, filled=True)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score ,classification_report
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
# And obviously you can't do this for a biger data set that's why you can go with 
# pri pruning
```

```python
# pri pruning  >> It is also called as hyperparameter turning
```

```python
# how to do it >> So i will define parameter  >> let see
```

```python
parameter = {
    'criterion': ['gini' , 'entropy','log_loss'],
    'max_depth': [1,2,3,4,6,8],
    'splitter' : ['best', 'random'],
    'max_features':['sqrt','log2'],   
} 
```

```python
from sklearn.model_selection import GridSearchCV
# GridSearchCV >>  helps find the best combination of hyperparameters for a model
```

```python
# now make a empty object of your decision tree classifier
# let  see
```

```python
clf =DecisionTreeClassifier()
clf
```

```python
# now pass this  object of your decision tree classifier to  GridSearchCV
```

```python
model= GridSearchCV(clf, param_grid=parameter, cv=5, scoring="accuracy")
model
```

```python
model.fit(X_train,y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```
