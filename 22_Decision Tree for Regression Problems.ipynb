```python
import numpy as numpy 
import pandas  as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")\


from sklearn.datasets import fetch_california_housing 
data=fetch_california_housing()
df= pd.DataFrame(data.data,columns=data.feature_names)
```

```python
df
```

```python
df['Price']= data.target
df
```

```python
df= df.sample(frac = 0.20)
```

```python
df
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=1)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize =(12, 10))
tree.plot_tree(model, filled=True)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
# Hyperparameter turning 
```

```python
parameter = {
    'criterion': ['squared' , 'poisson'],
    'max_depth': [1,2,3,4,6,8],
    'splitter' : ['best', 'random'],
    'max_features':['sqrt','log2'],   
} 
reg= DecisionTreeRegressor()
```

```python
from sklearn.model_selection import GridSearchCV
model=GridSearchCV(reg, param_grid= parameter , cv = 5 , scoring= "neg_mean_squared_error")
```

```python
model
```

```python
model.fit(X_train, y_train)
```

```python
model.best_params_
```

```python
final_model= DecisionTreeRegressor(criterion = "squared_error",max_depth=8)
```

```python
final_model.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize =(12, 10))
tree.plot_tree(final_model, filled=True)
```
