```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import make_regression
X, y = make_regression(n_samples=1000, n_features=2,n_targets=1, noise=0.3)
```

```python
X
```

```python
y # y is counties variable 
```

```python
sns.scatterplot(x = pd.DataFrame(X)[0], y= pd.DataFrame(X)[1],hue =y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30,random_state=1)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.svm import SVR
svr =SVR(kernel="linear")
```

```python
svr
```

```python
svr.fit(X_train,y_train)
```

```python
y_pred = svr.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
# hyperparameter tuning >> Grid search CV
```

```python
from sklearn.model_selection import GridSearchCV
params = {
    'C': [1,2,3,4,5,10,50,100], # C is number of miss classified data point.
    'gamma': [0.1,0.2,0.001],# gamma means scaling factors  >> any value
    'kernel': ['linear'],
    'epsilon':[0.1,0.001,0.2,0.3]
}
```

```python
params
```

```python
grid = GridSearchCV(SVR(),param_grid=params,cv=5,verbose=3)
```

```python
grid
```

What means verbose see it 

verbose=3 means that it give you intermediate results .

verbose=2 means that it give you overall result .

verbose=1 means that it give you how many fitted is their  by-default it will be.

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```
